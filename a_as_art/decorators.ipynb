Napisz dekorator `@wraps`, który będzie działał tak jak ten dostępny w bibliotece standardowej: `functoolds.wraps`.

```python
# solution

def wraps(to_be_decorated):
  def new_wrapper(*args, **kwargs):
    result = to_be_decorated
    result.__name__ = to_be_decorated.__name__
    result.__doc__ = to_be_decorated.__doc__
    result.__module__ = to_be_decorated.__module__
    result.__qualname__ = to_be_decorated.__qualname__
    result.__annotations__ = to_be_decorated.__annotations__
    return result
  return new_wrapper
```

```python
# usage

def bumelant(*args, **kwargs):
    def wrap(to_be_decorated):
        @wraps(to_be_decorated)
        def wrapper(*args, **kwargs):
            return to_be_decorated(*args, **kwargs)
        return wrapper
    return wrap


@bumelant(1, 2)
def say_python():
    """jakis opis"""
    return "PYTHON"


assert say_python.__doc__ == "jakis opis"
```

Napisz dekorator `@is_correct`,  który opakowuje funkcję zwracającą słownik. Dekorator ma sprawdzić czy w słowniku znajdują się klucze zawarte w argumentach dekoratora. Jeśli tak niech zwróci ten słownik, jeśli nie, niech zwraca wartość `None`.

```python
# solution

def is_correct(*args):
  def decorator(function):
    def wrapper():
      slownik = function()
      return slownik if all( item in slownik for item in args) else None
    return wrapper
  return decorator
```

```python
# usage

@is_correct('first_name', 'last_name')
def get_data():
    return {
        'first_name': 'Jan',
        'last_name': 'Kowalski',
        'email': 'jan@kowalski.com'
    }


@is_correct('first_name', 'last_name', 'email')
def get_other_data():
    return {
        'first_name': 'Jan',
        'email': 'jan@kowalski.com'
    }


assert get_data() == {
        'first_name': 'Jan',
        'last_name': 'Kowalski',
        'email': 'jan@kowalski.com'
    }


assert get_other_data() is None
```

Napisz dekorator `@add_date`,  który opakowuje funkcję zwracającą słownik. Dekorator ma dodać aktualną datę do zwracanego przez dekorowaną funkcję słownika w formacie podanym jako argument dekoratora.

Użyj modułu datetime korzystając z `datetime.datetime.now()` do pobrania aktualnej daty.

```python
# solution

import datetime

def add_date(format):
    def decorator(function):
        def wrapper(*args, **kwargs):
            now = datetime.datetime.now()
            result = function(*args, **kwargs)
            result['date'] = now.strftime(format)
            return result
        return wrapper
    return decorator
```

```python
# usage

@add_date('%B %Y')
def get_data(a):
    return {1: a, 'name': 'Jan'}


assert get_data(2) == {
    1: 2, 'name': 'Jan', 'date': 'April 2020'
}
```
